```python


df_grouped = data.groupby('State')

```

```python


max_males_vaccinated = df_grouped['Male(Individuals Vaccinated)'].max()['India']
max_males_vaccinated

```

```python
female_vaccinated = df_grouped['Female(Individuals Vaccinated)'].max()['India']
female_vaccinated
```

```python
first_dose = data.groupby('State')[['State','First Dose Administered']].max()
first_dose = first_dose.drop('India')
first_dose.sort_values(by='First Dose Administered', ascending=False, inplace=True)
first_dose
```

```python
second_dose = data.groupby('State')[['State', 'Second Dose Administered']].max()

second_dose = second_dose.drop('India')
second_dose.sort_values(by='Second Dose Administered', ascending=False, inplace=True)
second_dose
```

```python
sns.catplot(
    data=first_dose, y="First Dose Administered",x="State",
    kind="bar", height=40, aspect=2,
)
```

```python
sns.catplot(
    data=second_dose, y="Second Dose Administered",x="State",
    kind="bar", height=40, aspect=2,
)
```

```python



fig, ax = plt.subplots()


wedges, texts = ax.pie(first_dose['First Dose Administered'], labels=None)

ax.set_title('Pie Chart')

ax.legend(wedges, first_dose['State'],bbox_to_anchor=(-0.1,1))

plt.show()
```

```python
fig, ax = plt.subplots()


wedges, texts = ax.pie(second_dose['Second Dose Administered'], labels=None)

ax.set_title('Pie Chart')

ax.legend(wedges, second_dose['State'],bbox_to_anchor=(-0.1,1))

plt.show()
```

```python
IndiaState = data.query("State == 'India'")
ax = sns.lineplot(data=IndiaState, x="Updated On", y="Total Doses Administered", hue='State', palette='bright')
ax.set_xticklabels([])
plt.show()
```

```python
sns.set_style('darkgrid')
data1 = data.drop(data[data['State'] == 'India'].index)



fig, ax = plt.subplots(figsize=(12, 8))

sns.lineplot(data=data1, x='Updated On', y='Total Doses Administered', hue='State', palette='bright', ax=ax)

ax.set_xticklabels([])
ax.set_xlabel('')
ax.set_ylabel('Total Doses Administered')
ax.set_title('Total Doses Administered by State')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```
