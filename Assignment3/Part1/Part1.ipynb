```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('nba.csv')
```

```python
data.head()
```

```python
data['Salary'].isnull().sum()
```

```python
data['Age'].isnull().sum()
```

```python
data.dtypes
```

```python
df = data['Salary'].fillna(data['Salary'].mean())
data['Salary'] = df;
df = data['Age'].fillna(data['Age'].median())
data['Age'] = df
```

```python
data.dtypes
```

```python
data['Age'] = data['Age'].astype(int)
```

```python
data.dtypes
```

```python
data.groupby('Age')['Salary'].describe()
```

```python
data.groupby('Age')['Salary'].size()
```

```python
data.groupby('Age')['Salary'].agg(['mean','median','std','min','max'])
```

```python
data.groupby('Age')['Salary'].mean()
```

```python
data.groupby('Age')['Salary'].median()
```

```python
data.groupby('Age')['Salary'].std()
```
